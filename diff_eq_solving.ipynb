### euler method to solve diff. equation


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def f(y):
    return 1+y**2
    
def euler(f ,h, y ):
    return y+(h*f(y))

def solver(euler, initial, h, f):
    X = np.arange(0, 1.54+h, h)         #chose 1.54 because its less than pi/2
    Y = np.zeros(len(X))
    Y[0] = initial
    for i in range(len(X)-1):
        Y[i+1] = euler(f, h, Y[i])
    return X,Y
Xaxis,Yaxis = solver(euler, 0, 0.01, f) 
```

```python
plt.scatter(Xaxis, Yaxis);
y1 = np.tan(X)
plt.scatter(Xaxis,y1, label = "tanx")
plt.legend()
```

## modified euler method

```python
def modified_solver(euler, initial, h, f):
    X = np.arange(0, 1.54+h, h)
    Y = np.zeros(len(X))
    Y[0] = initial
    for i in range(len(X)-1):
        yt = euler(f, h, Y[i])
        ft = (f(yt)+f(Y[i]))/2
        Y[i+1] = Y[i] + h*ft
    return X,Y
Xaxis1, Yaxis1 = modified_solver(euler, 0, 0.01, f)
```

```python
plt.scatter(Xaxis, Yaxis, marker=".",label = "euler method solution");
y1 = np.tan(X)
plt.scatter(Xaxis,y1, marker=".", label = "tanx");
plt.legend()
plt.scatter(Xaxis1, Yaxis1, marker=".", label = "modified euler solution" );
plt.legend()
```
