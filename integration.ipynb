## Monte Carlo Intigration

```python

count = 0
iter = 10000
X,Y = np.zeros(iter) , np.zeros(iter)
for _ in range(iter):
    x = np.random.rand()*np.pi
    y = np.random.rand()
    X[_] = x
    Y[_] = y
    if ( y < np.sin(x)**3 ):
        count += 1
intigration = (count/iter)*np.pi
print(intigration)
```

```python
x = np.linspace(0,np.pi, 10001)
y = np.sin(x)**3

plt.scatter(x,y, color = "g" ,marker = ".")

for _ in range(iter):
    if ( Y[_] < np.sin(X[_])**3 ):
        plt.scatter(X[_],Y[_], color = "red" ,marker = ".")
    else:
        plt.scatter(X[_],Y[_] ,color = "blue", marker = ".")
```
